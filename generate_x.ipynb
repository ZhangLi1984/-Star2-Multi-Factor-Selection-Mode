### generate x

```python
import pickle
def load_obj(name):
    with open(name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
# load x data
dict1 = load_obj("data_store/factor_data1_1")
dict2 = load_obj("data_store/factor_data1_2")
dict3 = load_obj("data_store/factor_data1_3")
```

```python
# get the attr
attr_list = list(dict1.keys())
```

```python
dict2["AR"]
```

```python
# merge the dict, actually, 300*1400 = 420,000, but there is null, so smaller
attr_dict = {}
for i in tqdm(attr_list[:]):
    tmp1 = dict1[i].groupby(dict1[i].index).first()
    tmp2 = dict2[i].groupby(dict2[i].index).first()
    tmp3 = dict3[i].groupby(dict3[i].index).first()
    tmp123 = pd.concat([tmp1,tmp2,tmp3])
    tmp123.index = tmp123.index.astype("str")
    tmp123 = tmp123.groupby(tmp123.index).first()
    # use stack func to reshape the df
    tmp = tmp123.stack().reset_index()
    tmp.columns = ["date","code",i]
    # sort by (code,date)
    tmp = tmp.sort_values(by = ["code","date"],ascending = True).reset_index(drop=True)
    attr_dict[i] = tmp
```

```python
# show,300*1400 = 420,000, but there is null, so smaller
attr_dict["AR"]
```

```python
# transfer dict to a big df
df = attr_dict[attr_list[0]]
for i in tqdm(attr_list[1:]):
    df = pd.merge(df,attr_dict[i],on=["date","code"],how="outer")
```
